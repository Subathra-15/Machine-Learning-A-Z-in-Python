Importing the libraries

```python
import numpy as np #for arrays
import matplotlib.pyplot as plt #for visualization
import pandas as pd #preprocess and import datasets
```

Importing the dataset

```python
data=pd.read_csv("Data.csv") #creates a dataframe
```

Features(independent variable) and predict(dependent variable)

Seperate different matrix for both

```python
X=data.iloc[:,:-1].values #locate indexes. columns and rows
y=data.iloc[:,-1].values
print(X)
print(y)
#matrix feature and dependent variable vector
```

Missing datas

```python
#use scikit learn to handle missing data
#replace missing with average of the other values
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
#fit method look at missing values and compute the average
#transform method will be applied to the datas
imputer.fit(X[:,1:3])
X[:,1:3]=imputer.transform(X[:,1:3])
print(X)


```

Encoding the categorical data

```python
#turn strings into numbers
#one hot encoding. turning one column into n diff columns
```

Encoding the independent variable

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
#create object
# 1st column will be the transformer column and 2nd one will be the remaining columns
#pass through keeps the remaining column intact
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough')
X=np.array(ct.fit_transform(X))
# X has to be a numpy array
print(X)
```

Encoding the dependent variable

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y) #doesnt have to be a numpy array
print(y)
```

Split the dataset into training and testing

```python
from sklearn.model_selection import train_test_split
#80% train set and 20% test set
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
print(X_train)
print(X_test)
print(y_train)
print(y_test)
```

Feature Scaling

```python
#not applied to all the models
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train[:,3:]=sc.fit_transform(X_train[:,3:])
X_test[:,3:]=sc.transform(X_test[:,3:])
print(X_train)
print(X_test)
```
